```python
from bs4 import BeautifulSoup
import requests
from os import setgroups
```

```python
url = "https://www.rmg.co.uk/stories/topics/solar-system-data"
response = requests.get(url)  # Get the webpage content
page = response.text  # Now 'page' is defined
soup = BeautifulSoup(page, 'html')
```

```python
table = soup.find_all('table')[0]
```

```python
rows = table.find_all('tr')
```

```python
headers = [header.get_text(strip=True) for header in rows[0].find_all('p')]
```

```python
# Extract data rows
data = []
for row in rows[1:]:  # Skip header row
    values = [cell.get_text(strip=True) for cell in row.find_all('p')]
    data.append(values)
```

```python
import pandas as pd
```

```python
# Create DataFrame
df = pd.DataFrame(data, columns=headers)
```

```python
df
```

```python
# Save to CSV
csv_filename = "solar_system_data.csv"
df.to_csv(csv_filename, index=False)

print(f"CSV file '{csv_filename}' has been created successfully!")
```
